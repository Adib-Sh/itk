```python
def findValue(dct, value):
    if isinstance(dct, dict):
        for key in dct:
            if dct[key] == value:
                return key
            found = findValue(dct[key], value)
            if found:
                found.append(key)
                return found
        return None
    elif isinstance(dct, list):
        for i, item in enumerate(dct):
            if isinstance (item, dict) or isinstance(item, list):
                found = findValue(item, value)
                if found:
                    found.append(i)
                    return found
            else:
                if item == value:
                    return i
        return None
```
