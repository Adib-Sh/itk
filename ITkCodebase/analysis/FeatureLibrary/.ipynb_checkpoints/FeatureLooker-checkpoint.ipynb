```python
def getFScores(chip, stream, variable):
    channels = [channel for channel in chip.getStream(stream) if not channel.badchannel]
    leftChannels = channels[:len(channels)//2]
    rightChannels = channels[len(channels)//2:]
    mean = {"outnse" : chip.meanOutnse,
            "innse"  : chip.meanInnse,
            "gain"   : chip.meanGain}[variable](stream=stream, countbadchannels=False)
    sLeft = 0
    sRight = 0

    values = [{"outnse" : channel.outnse,
               "innse"  : channel.innse,
               "gain"   : channel.gain}[variable] for channel in leftChannels[-1:0:-1]]
    xs = [x / len(leftChannels) for x in range(len(leftChannels))]
    ys = [y - mean for y in values]
    yMean = np.mean(np.abs(ys))
    ys = np.array(ys) / yMean
    ws = [x * y for x, y in zip(xs, ys)]
    sLeft = sum(ws)
    
    values = [{"outnse" : channel.outnse,
               "innse"  : channel.innse,
               "gain"   : channel.gain}[variable] for channel in rightChannels[-1:0:-1]]
    xs = [x / len(rightChannels) for x in range(len(rightChannels))]
    ys = [y - mean for y in values]
    yMean = np.mean(np.abs(ys))
    ys = np.array(ys) / yMean
    ws = [x * y for x, y in zip(xs, ys)]
    sRight = sum(ws)

    return (sLeft, sRight)
```

```python
def getGradient(vector, step):
    grad = [None] * step
    for p1, p2 in zip(vector[:-2*step], vector[step*2:]):
        grad.append((p2 - p1) / (2 * step))
    grad.extend([None] * step)
    gradSum = 0
    for element in grad:
        if element is not None:
            gradSum += np.abs(element)
    return grad, gradSum
```

sum(x_i * [y_i - M]) = sum(x_i * [k * x_i + m - M]) = sum(x_i * [k * x_i + c]) = sum(k * x²_i + c * x_i)
-> integral limit -> = k * x³_i / 3 + c * x²_i / 2

```python
chipScores = []
gradientStep = 5

for hybrid in featuredHybrids:
    hybrid.classification()
    for chip in hybrid.chips:
        FLeft, FRight = getFScores(chip, 0, "outnse")
        slopeScore = FLeft - FRight
        skewScore = FLeft + FRight
        values = [channel.outnse for channel in chip.getStream(0) if not channel.badchannel]
        _, gradSum = getGradient(values, gradientStep)
        chipScores.append((chip, slopeScore, skewScore, gradSum))
maxSlope = -1000000
minSlope = 1000000
maxSkew = -1000000
minSkew = 1000000
maxGrad = -1000000
minGrad = 1000000
for listing in chipScores:
    maxSlope = max(maxSlope, listing[1])
    minSlope = min(minSlope, listing[1])
    maxSkew = max(maxSkew, listing[2])
    minSkew = min(minSkew, listing[2])
    maxGrad = max(maxGrad, listing[3])
    minGrad = min(minGrad, listing[3])
    
nBins = 14

slopeEdges = [np.ceil(minSlope + (i + 1) * (maxSlope - minSlope) / nBins) for i in range(nBins)]
slopeEdges.sort()
skewEdges = [np.ceil(minSkew + (i + 1) * (maxSkew - minSkew) / nBins) for i in range(nBins)]
skewEdges.sort()
gradEdges = [np.round(minGrad + (i + 1) * (maxGrad - minGrad) / nBins, 2) for i in range(nBins)]
gradEdges.sort()
chipBins = {"slope" : {},
            "skew" : {},
            "grad" : {}}

done = False
for listing in chipScores:
    for edge in slopeEdges:
        if edge not in chipBins["slope"].keys():
            chipBins["slope"][edge] = []
        if listing[1] < edge:
            chipBins["slope"][edge].append(listing[0])
            break
    for edge in skewEdges:
        if edge not in chipBins["skew"].keys():
            chipBins["skew"][edge] = []
        if listing[2] < edge:
            chipBins["skew"][edge].append(listing[0])
            break
    for edge in gradEdges:
        if edge not in chipBins["grad"].keys():
            chipBins["grad"][edge] = []
        if listing[3] < edge:
            chipBins["grad"][edge].append(listing[0])
            break
```

```python
maxStack = 0
for scoreType in ["slope", "skew", "grad"]:
    for slopeBin in chipBins[scoreType]:
        maxStack = max(maxStack, len(chipBins[scoreType][slopeBin]))

fig = plt.figure(figsize=(10, maxStack))
        
maxRange = 0
badChips = []
for i, slopeBin in enumerate(chipBins["slope"]):
    for c, chip in enumerate(chipBins["slope"][slopeBin]):
        values = np.array([channel.outnse for channel in chip.getStream(0) if not channel.badchannel])
        if len(values) == 0:
            badChips.append(c)
        else:
            maxRange = max(maxRange, values.max()-values.min())
        
for i, slopeBin in enumerate(chipBins["slope"]):
    stackHeight = 0
    for c, chip in enumerate(chipBins["slope"][slopeBin]):
        if c in badChips:
            continue
        values = np.array([channel.outnse for channel in chip.getStream(0) if not channel.badchannel])        
        thisRange = values.max() - values.min()
        values = ((values - values.min()) / (values.max() - values.min())) * thisRange/maxRange + stackHeight
        stackHeight += 1
        plt.scatter([128*i + x for x in range(len(values))], values, s=2)
plt.xticks([128*i for i in range(nBins)]);
plt.gca().set_xticklabels(list(chipBins["slope"].keys()));
plt.ylim([0, maxStack])
plt.yticks([i for i in range(maxStack)])
plt.suptitle("Slope (leftScore - rightScore)")
plt.grid()
```

```python
# Build bins using all three scores; slope, skew, grad
variables = ["slope", "skew", "grad"]

xVariable = "grad"
yVariable = "skew"
colorVariable = "slope"

listingDict = {"slope" : 1, "skew" : 2, "grad" : 3}

maxValues = {"slope" : -1000000, "skew" : -1000000, "grad" : -1000000}
minValues = {"slope" : 1000000, "skew" : 1000000, "grad" : 1000000}
allValues = dict(zip(variables, [[]] * 3))

for listing in chipScores:
    for variable in variables:
        maxValues[variable] = max(maxValues[variable], listing[listingDict[variable]])
        minValues[variable] = min(minValues[variable], listing[listingDict[variable]])
        allValues[variable].append(listing[listingDict[variable]])
        
for variable in variables:
    allValues[variable].sort()
        
nBins = 12
nDecimals = {"slope" : 1, "skew" : 1, "grad" : 2}
binEdges = np.round(np.linspace(minValues[xVariable], maxValues[xVariable], nBins+1)[1:], nDecimals[xVariable])
maxValueRange = -1000000

# Build xVariable bins
chipBins = {}
for listing in chipScores:
    
    chip = listing[0]
    values = [channel.outnse for channel in chip.getStream(0) if not channel.badchannel]
    maxValueRange = max(maxValueRange,
                        np.max(values) - np.min(values))
    
    for edge in binEdges:
        if edge not in chipBins.keys():
            chipBins[edge] = []
        if listing[listingDict[xVariable]] < edge:
            chipBins[edge].append(listing)
            break

# Sort along y axis
maxStack = 0
for edge in binEdges:
    chipBins[edge].sort(key = lambda x : x[listingDict[yVariable]])
    maxStack = max(maxStack, len(chipBins[edge]))

colors = [(i*1/nBins, 0, 1-(i*1/nBins)) for i in range(nBins)]

fig = plt.figure(figsize=(12, 8))
for i, xBin in enumerate(chipBins):
    binList = chipBins[xBin]
    minDist = 1000000
    midIdx = -1
    binMax = -1000000
    binMin = 1000000
    mid = maxValues[yVariable] - (maxValues[yVariable] - minValues[yVariable]) / 2
    for l, listing in enumerate(binList):
        dist = np.abs(listing[listingDict[yVariable]] - mid)
        if dist < minDist:
            minDist = dist
            midIdx = l
        binMax = max(binMax, listing[listingDict[yVariable]])
        binMin = min(binMin, listing[listingDict[yVariable]])
    
   
    
    stackHeight = len(binList)
    maxRange = maxValues[yVariable] - minValues[yVariable]
    thisRange = binMax - binMin
    step = 1 # maxRange / (maxStack - 1)
    startPos = -midIdx
    
    #print("In bin", i, ": stackheight", stackHeight, "maxRange", maxRange, "thisRange", thisRange, "startPos", startPos)
    
    for c, listing in enumerate(chipBins[xBin]):
        chip = listing[0]
        values = np.array([channel.outnse for channel in chip.getStream(0) if not channel.badchannel])
        thisValueRange = values.max() - values.min()
        values = ((values - values.min()) / (values.max() - values.min())
                 ) * thisValueRange/maxValueRange + c + startPos
        color = ((listing[listingDict[colorVariable]] - minValues[colorVariable]) / (maxValues[colorVariable] - minValues[colorVariable]),
                 0, 1-(listing[listingDict[colorVariable]] - minValues[colorVariable]) / (maxValues[colorVariable] - minValues[colorVariable]))
        plt.scatter([128*i + x for x in range(len(values))], values, s=2, color=color)
plt.xticks([128*i for i in range(nBins)]);
plt.gca().set_xticklabels(list(chipBins.keys()));
plt.yticks([i for i in range(int(plt.ylim()[0]), int(plt.ylim()[1]))])
plt.suptitle("Chip score chart")
plt.grid()
plt.xlabel(xVariable + " score")
plt.ylabel(yVariable + " bin [a.u.]")
```
